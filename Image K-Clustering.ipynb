```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
import matplotlib.image as mpimg
```

```python
data=mpimg.imread("scene.jpg")
```

```python
plt.imshow(data)
```

```python
(h,w,c)=data.shape
```

```python
image=data.reshape(h*w,c)
```

```python
km= KMeans(n_clusters=8)
```

```python
labels=km.fit_predict(image)
```

```python
rbf=km.cluster_centers_
```

```python
image=np.round(rbf[labels],0).astype(int)
```

```python
image=image.reshape(h,w,c)
```

```python
plt.imshow(image)
```
